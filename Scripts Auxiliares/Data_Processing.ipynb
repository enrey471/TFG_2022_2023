```python
reference = []
with open('reference_wavelengths.txt') as file:
    for line in file:
        reference.append(line[3:11])
```

```python
reference
```

```python
select = []
with open('splib07a_Wavelengths_ASD_0.35-2.5_microns_2151_ch.txt') as file:
    for line in file:
        select.append(line[1:15])
```

```python
select = select[1:len(select)]
select
```

```python
values = []
with open('splib07a_Serpentine_HS318.1B_ASDFRc_AREF.txt') as file:
    for line in file:
        values.append(line[1:15])
```

```python
values = values[1:len(values)]
values
```

```python
result = []
j = 0 # indice (data) de la posición actual
for i in range(0,224):
    found = False
    while not found:
        if (abs(float(select[j]) * 1000 - float(reference[i]))  < abs((float(select[j + 1]) * 1000 - float(reference[i])))):
            found = True 
            result.append(float(values[j]) * 1000)
        j = j + 1
```

```python
len(result)
```

```python
resultString = ','.join(map(str, result))
```

```python
resultString
```

```python
text_file = open("A_Serpentine.txt", "w")
text_file.write(resultString)
text_file.close()
```
